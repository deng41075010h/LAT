```python
#載入csv
import pandas as pd

df = pd.read_csv('8f8f0105-b5bb-424e-9921-acd69358eda6.csv')

df
```

```python
import plotly.graph_objects as go

# 繪製堆疊條形圖
fig = go.Figure()
for department in df['系所名稱'].unique():
    fig.add_trace(go.Bar(
        x=df[df['系所名稱'] == department]['校名'],
        y=df[df['系所名稱'] == department]['總修課人數'],
        name=department
    ))

fig.update_layout(
    title='各學校遠距課程修課學生人數',
    xaxis_title='學校名稱',
    yaxis_title='修課人數',
    barmode='stack'
)

fig
#jupyter上有圖，但上傳github就沒有，因此另外附圖
```

Q1：由圖可知，哪幾個學校有最多的遠距課程選修人數？

Ans：銘傳大學、淡江大學、中原大學，其中可以很明顯看出中原大學的人文教育學院所開設的課程擁有非常多選修人數

Q2：哪一個系所所開設的課程有相對多的選修人數？

Ans：除了中原大學等本身選修人數就多的學校外，可以發現成功大學雖然總選修人數不多，但是很集中在外國語言及通識中心所開設的課程；
另外台北醫學大學也集中在一般通識組所開設的課程

Q3：哪一學校可能較有多元豐富的遠距課程提供給學生選修？

Ans：由圖可發現，除了銘傳大學有非常多選修人數外，台灣大學、亞洲大學及長榮大學雖然總選修人數相對沒那麼多，但選修的課程卻仍非常多元

想把第1學期及第2學期拆分出來作圖......

```python
# 設定圖表
fig = go.Figure()

# 第1學期
fig.add_trace(go.Bar(
    x=df.loc[df['開課學期']==1, '校名'],
    y=df.loc[df['開課學期']==1, '總修課人數'],
    name='第1學期',
    text=df.loc[df['開課學期']==1, '系所名稱'],
    hovertemplate='校名: %{x}<br>總修課人數: %{y}<br>系所名稱: %{text}',
    offsetgroup=0,
    #offset = 0
))

# 第2學期
fig.add_trace(go.Bar(
    x=df.loc[df['開課學期']==2, '校名'],
    y=df.loc[df['開課學期']==2, '總修課人數'],
    name='第2學期',
    text=df.loc[df['開課學期']==2, '系所名稱'],
    hovertemplate='校名: %{x}<br>總修課人數: %{y}<br>系所名稱: %{text}',
    offsetgroup=0,
    #offset = 1
))

# 設定圖表樣式
fig.update_layout(
    barmode='relative',
    xaxis={'title': '學校'},
    yaxis={'title': '修課人數'},
    title='學校修課人數統計'
)

# 顯示圖表
fig.show()
```

覺得上面資料太雜亂，改取前15名

```python
# 取前15名的學校
top15 = df.groupby('校名').sum().sort_values('總修課人數', ascending=False).iloc[:15]
top15
```

```python
# 計算每個學校的總修課人數
df_total = df.groupby('校名', as_index=False)['總修課人數'].sum()

# 以總修課人數排序
df_total = df_total.sort_values('總修課人數', ascending=False)

# 取前15名學校
top_15_schools = df_total['校名'].iloc[:15].tolist()

# 篩選出前15名學校的資料
df_top_15 = df[df['校名'].isin(top_15_schools)]

# 轉換資料結構，以學校和學期為索引
df_pivot = df_top_15.pivot_table(index=['校名', '開課學期'], columns='系所名稱', values='總修課人數', fill_value=0)

# 設定圖表
fig = go.Figure()

# 第1學期
fig.add_trace(go.Bar(
    x=df_pivot.loc[(slice(None), 1), :].index.get_level_values('校名'),
    y=df_pivot.loc[(slice(None), 1), :].sum(axis=1),
    name='第1學期',
    offsetgroup=0
))

# 第2學期
fig.add_trace(go.Bar(
    x=df_pivot.loc[(slice(None), 2), :].index.get_level_values('校名'),
    y=df_pivot.loc[(slice(None), 2), :].sum(axis=1),
    name='第2學期',
    offsetgroup=1
))

# 設定圖表樣式
fig.update_layout(
    barmode='relative',
    xaxis={'title': '學校'},
    yaxis={'title': '修課人數'},
    title='前15名學校修課人數統計'
)

# 顯示圖表
fig.show()
```
