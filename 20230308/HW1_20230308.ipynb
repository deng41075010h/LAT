Q1 : 哪一個領域(系所)的數位課程修課人數最多？

```python
temp1 = df[['系所名稱', '總修課人數']]
Type1 = temp1.groupby(['系所名稱'])['總修課人數'].sum().reset_index()
Type1 = Type1.sort_values(by='總修課人數', ascending=False)
Type1
```

延伸Q7
開課數量與修課總人數在不同學校的比較與關係

```python
# Group by school and calculate total number of people and courses
semester_totals2 = df.groupby('校名').agg({'課程名稱': 'nunique', '總修課人數': 'sum'})
# Change column title
semester_totals2 = semester_totals2.rename(columns={'課程名稱': '開課數量'})
semester_totals2
```

計算相關係數

```python
correlation = semester_totals2['開課數量'].corr(semester_totals2['總修課人數'])
correlation
```

為高度正相關 存在很強的線性關係

```python
# Calculate average number of people per course
semester_totals2['平均修課人數'] = semester_totals2['總修課人數'] / semester_totals2['開課數量']

semester_totals2.sort_values(by='平均修課人數', ascending=False)
```
