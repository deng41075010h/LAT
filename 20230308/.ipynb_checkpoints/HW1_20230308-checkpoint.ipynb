Q1 : 哪一個領域(系所)的數位課程修課人數最多？

Q10 : 哪一個學校的哪門數位課程有最多選修人數？

Q11 : 開課數量與修課總人數在不同學期的比較與關係

```python
# Group by semester and calculate total number of people and courses
semester_totals = df.groupby('開課學期').agg({'課程名稱': 'nunique', '總修課人數': 'sum'})
# Change column title
semester_totals = semester_totals.rename(columns={'課程名稱': '開課數量'})
# Display the results
semester_totals
```

計算相關係數

```python
correlation = semester_totals['開課數量'].corr(semester_totals['總修課人數'])
correlation
```

相關係數為1.0代表存在很強的線性關係

Q12 : 開課數量與修課總人數在不同學校的比較與關係

```python
# Group by school and calculate total number of people and courses
semester_totals2 = df.groupby('校名').agg({'課程名稱': 'nunique', '總修課人數': 'sum'})
# Change column title
semester_totals2 = semester_totals2.rename(columns={'課程名稱': '開課數量'})
semester_totals2
```

```python
correlation2 = semester_totals2['開課數量'].corr(semester_totals2['總修課人數'])
correlation2
```

```python
# Calculate average number of people per course
semester_totals2['人均修課人數'] = semester_totals2['總修課人數'] / semester_totals2['開課數量']

semester_totals2
```
