```python
import pandas as pd

df11201 = pd.read_csv('教育部國民及學前教育署112年1月補助明細表.csv')

df11201
```

```python
df11112 = pd.read_csv('教育部國民及學前教育署111年12月補助明細表.csv')

df11112.head()
```

```python
df11111 = pd.read_csv('教育部國民及學前教育署111年11月補助明細表.csv')

df11111.head()
```

```python
#第一種分析
#項目與金額
```

```python
temp = df11201[['科目名稱', ' 金額 ']]
temp
```

```python
Type = temp.groupby(['科目名稱'])[' 金額 '].sum().reset_index()
```

```python
Type['date'] = 11201
Type
```

```python
Type2 = df11112[['科目名稱', ' 金額 ']].groupby(['科目名稱'])[' 金額 '].sum().reset_index()
Type2['date'] = 11112
Type2
```

```python
Type3 = df11111[['科目名稱', ' 金額 ']].groupby(['科目名稱'])[' 金額 '].sum().reset_index()
Type3['date'] = 11111
Type3
```

```python
frames = [Type, Type2, Type3]

result = pd.concat(frames)

result
```

```python
result = result.reset_index()
result
```

```python
del result['index']
result
```

```python
#第二種分析
#受款單位與金額
```

```python
temp2 = df11201[['受款單位', ' 金額 ']]
temp2
```

```python
New = temp2.groupby(['受款單位'])[' 金額 '].sum().reset_index()
New['date'] = 11201
New
```

```python
New2 = df11112[['受款單位', ' 金額 ']].groupby(['受款單位'])[' 金額 '].sum().reset_index()
New2['date'] = 11112
New2
```

```python
New3 = df11112[['受款單位', ' 金額 ']].groupby(['受款單位'])[' 金額 '].sum().reset_index()
New3['date'] = 11111
New3
```

```python
frames2 = [New, New2, New3]

result2 = pd.concat(frames2)

result2
```
